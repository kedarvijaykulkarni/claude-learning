```python
# Load env veriables
from dotenv import load_dotenv

load_dotenv()
```

```python
# Create an API client
from anthropic import Anthropic

client = Anthropic()

model = "claude-sonnet-4-5"
```

```python
# functions 

# Helper functions to manage messages
def add_user_message(messages, text):
    messages.append({"role": "user", "content": text})

# Helper functions to manage messages
def add_assistant_message(messages, text):
    messages.append({"role": "assistant", "content": text})

# Function to send messages to the API and get a response
def chat(messages):
    response = client.messages.create(
        model=model,
        max_tokens=1000,
        messages=messages,
    )
    return response.content[0].text
```

```python
# Make an initial message list of messages
messages = []

# Use a While loop to have a conversation with the assistant
while True:
    user_input = input("You: ")
    add_user_message(messages, user_input)

    chat_response = chat(messages)
    print(f"Assistant: {chat_response}")

    add_assistant_message(messages, chat_response)
```
